```python
def handle(col):
    if ('高数' in col or '数分' in col or '微积分' in col) and '4.请问您所在的系是？' not in col:
        return 0
    elif ('计算机' in col or '编程' in col) and '4.请问您所在的系是？' not in col:
        return 1
    else:
        return 2
feature_types = list(map(handle, list(df.columns)))
feature_types
```
