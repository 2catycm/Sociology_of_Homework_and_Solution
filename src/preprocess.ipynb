```python
import pandas as pd
df = pd.read_excel("../data/data_20230527.xlsx")
df.head()
```

```python
df.dtypes
df.columns
```

```python
# 定性数据和定量数据
# 单选数据，多选数据，定量数据. 0, 1, 2
feature_types = [0, 0, 0, 0,
                 0, 0, 
                 *([1]*5),*([1]*3),
                 0, 0,
                 0, 0,
                 *([1]*5),*([1]*3),
                 0, 
                 *([2]*4*2),
                 *([1]*8)
                 ]
assert len(feature_types)==len(df.columns)
len(feature_types)
```

```python
# 处理思路
# 单选：dummy化
# 多选：存在为True，不存在为Falses
# 定量：不存在为中位数。
for i, col in enumerate(df.columns):
    type = feature_types[i]
    if type==0:
        df = df.join(pd.get_dummies(df[col], prefix=col, prefix_sep="(单选)"))
        df = df.drop(col, axis=1)
    elif type==1:
        df[col] = df[col].apply(lambda x: not pd.isnull(x))
    elif type==2:
        df[col] = df[col].fillna(df[col].median())
```

```python
df.head()
df.to_excel("../result/preprocess_data.xlsx", index=False)
```

```python
# 另一种处理
for i, col in enumerate(df.columns):
    type = feature_types[i]
    if type==0:
        pass
    elif type==1:
        df[col] = df[col].apply(lambda x: not pd.isnull(x))
    elif type==2:
        df[col] = df[col].fillna(df[col].median())
df.head()
df.to_excel("../result/preprocess_data_no_dummy.xlsx", index=False)
```

```python
#针对连续数据分析
continous_cols = [i for i, type in enumerate(feature_types) if type==2]
continous_cols = df.columns[continous_cols]
continous_cols
continuous_df = df[continous_cols]
continuous_df.to_csv("../result/continuous_df.csv", index=False)
```
